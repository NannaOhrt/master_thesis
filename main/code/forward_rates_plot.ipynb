```python
import matplotlib.pyplot as plt
#data from citi 21.01.2024
forward_rates = [0.293838, 0.297602, 0.299622, 0.299258, 0.294395, 0.284018, 0.276305, 0.265454, 0.251744, 0.23401]
tenors = ['1Y', '2Y', '3Y', '5Y', '7Y', '10Y', '12Y', '15Y', '20Y', '30Y']
plt.figure(figsize=(14, 7))
plt.plot(tenors, forward_rates,  linestyle='--', color = 'tab:blue')
plt.plot(tenors, forward_rates, 'o', label ='Forward Rates')
plt.title("Forward Rates Over Different Tenors")
plt.xlabel("Tenor (years)")
plt.ylabel("Forward Rate (%)")
plt.grid(axis='y', linestyle='--', linewidth=0.5)
plt.legend()
plt.xticks(tenors)  
plt.savefig("/Users/nannaingemannohrt/Desktop/master_thesis/main/plots/forward_rates.png")
plt.show()
plt.close()

```
