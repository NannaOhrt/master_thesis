```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

# Download S&P 500 data from Yahoo Finance
sp500 = yf.download('^GSPC', start='1993-01-01', end='2023-01-01')

# Calculate daily returns
sp500['Returns'] = sp500['Adj Close'].pct_change()

# Calculate cumulative returns
sp500['Cumulative Returns'] = (1 + sp500['Returns']).cumprod()

# Calculate excess returns (subtract risk-free rate if available)
# For simplicity, assuming risk-free rate is 0 in this example
sp500['Excess Returns'] = sp500['Cumulative Returns'] - 1

# Calculate drawdown
sp500['Drawdown'] = 1 - sp500['Cumulative Returns'] / sp500['Cumulative Returns'].cummax()

# Plot cumulative excess returns with drawdowns highlighted
plt.figure(figsize=(10, 6))
plt.plot(sp500['Excess Returns'], label='Cumulative Excess Returns', color='blue')

# Highlight drawdowns as grey areas
drawdowns = sp500['Drawdown'] > 0
plt.fill_between(sp500.index, sp500['Excess Returns'], where=drawdowns, color='grey', alpha=0.3, label='Drawdown')

plt.title('Cumulative Excess Returns with Drawdowns Highlighted')
plt.xlabel('Year')
plt.ylabel('Cumulative Excess Returns')
plt.legend()
plt.show()
```

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

# List of tickers you want to compare
tickers = ['^GSPC', '^TYX', '^FVX']  # S&P 500, 30-Year Treasury Yield, 5-Year Treasury Note Yield

# Start and end dates
start_date = '1993-01-01'
end_date = '2023-01-01'

plt.figure(figsize=(12, 8))

for ticker in tickers:
    # Download data
    data = yf.download(ticker, start=start_date, end=end_date)
    
    # Calculate daily returns
    data['Returns'] = data['Adj Close'].pct_change()
    
    # Calculate cumulative returns
    data['Cumulative Returns'] = (1 + data['Returns']).cumprod()
    
    # Calculate excess returns (assuming risk-free rate is 0 for simplicity)
    data['Excess Returns'] = data['Cumulative Returns'] - 1
    
    # Plot
    plt.plot(data.index, data['Excess Returns'], label=f'{ticker} Cumulative Excess Returns')

# Adjust plot properties
plt.title('Cumulative Excess Returns for Different Tickers', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Cumulative Excess Returns', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

# Tickers for equities (S&P 500), nominal bonds (AGG), inflation-linked bonds (TIP), and 10-Year Treasury Yield (^TNX)
tickers = ['^GSPC', 'AGG', 'TIP', '^TNX']

# Start and end dates for the data
start_date = '1993-01-01'
end_date = '2023-01-01'

plt.figure(figsize=(12, 8))

for ticker in tickers:
    # Download historical data for the ticker
    data = yf.download(ticker, start=start_date, end=end_date)
    
    # Calculate daily returns
    data['Returns'] = data['Adj Close'].pct_change()
    
    # Calculate cumulative returns
    data['Cumulative Returns'] = (1 + data['Returns']).cumprod()
    
    # Normalize starting point to 100 for comparison
    initial_value = data['Cumulative Returns'].dropna().iloc[0]
    data['Normalized Returns'] = data['Cumulative Returns'] / initial_value * 100
    
    # Plot the normalized cumulative returns for each ticker
    plt.plot(data.index, data['Normalized Returns'], label=f'{ticker} Normalized Cumulative Returns')

# Adjust plot settings for better visualization
plt.title('Comparison of Normalized Cumulative Returns', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Normalized Cumulative Returns', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
