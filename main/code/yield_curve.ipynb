```python
import pandas as pd
import matplotlib.pyplot as plt

# Data from Citi
data = [5.33989, 5.33989, 5.33107, 5.21014, 5.06144, 4.95308, 4.83148, 4.52469, 4.35589, 4.17572]

# Define terms
terms = ['1M', '3M', '6M', '1Y', '2Y', '3Y', '5Y', '10Y', '20Y', '30Y']

# Create DataFrame
df = pd.DataFrame(data, columns=['Yield'], index=terms)

# Plot the yield curve
plt.figure(figsize=(10, 6))
plt.plot(terms, df['Yield'], marker='o', linestyle='-')
plt.title('Yield Curve')
plt.xlabel('Term (Years)')
plt.ylabel('Yield')
plt.grid(True)
plt.xticks(terms, terms)
plt.show()
```
