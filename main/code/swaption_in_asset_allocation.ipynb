```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

# Tickers for European equities (FEZ), nominal bonds (AGG), inflation-linked bonds (TIP), and 10-Year Treasury Yield (^TNX)
tickers = ['FEZ', 'AGG', 'TIP', '^TNX']

# Start and end dates for the data
start_date = '2004-01-01'
end_date = '2024-01-01'

plt.figure(figsize=(12, 8))

for ticker in tickers:
    # Download historical data for the ticker
    data = yf.download(ticker, start=start_date, end=end_date)
    
    # Calculate daily returns
    data['Returns'] = data['Adj Close'].pct_change()
    
    # Assuming risk-free rate is 0, so excess returns are equal to actual returns in this case
    data['Excess Returns'] = data['Returns']
    
    # Calculate cumulative excess returns
    data['Cumulative Excess Returns'] = (1 + data['Excess Returns']).cumprod()
    
    # Normalize starting point to 100 for comparison
    initial_value = data['Cumulative Excess Returns'].dropna().iloc[0]
    data['Normalized Cumulative Excess Returns'] = data['Cumulative Excess Returns'] / initial_value * 100
    
    # Plot the normalized cumulative excess returns for each ticker
    plt.plot(data.index, data['Normalized Cumulative Excess Returns'], label=f'{ticker} Normalized Cumulative Excess Returns')

# Adjust plot settings for better visualization
plt.title('Comparison of Normalized Cumulative Excess Returns', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Normalized Cumulative Excess Returns', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

```
