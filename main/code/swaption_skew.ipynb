```python
# Extract the '10Y10Y' row from the original dataframe
ten_y_1_y_2023 = df_OTM_2023.loc['10Y1Y']
ten_y_1_y_2024 = df_OTM_2024.loc['10Y1Y']

# Convert indices to integer deltas (spread measure in bps)
OTM_10Y1Y_2023 = ten_y_1_y_2023.index.str.replace('ATM', '0').astype(int)
OTM_10Y1Y_2024 = ten_y_1_y_2024.index.str.replace('ATM', '0').astype(int)


plt.figure(figsize=(10, 6))
plt.plot(OTM_10Y1Y_2023, ten_y_1_y_2023.values, color='blue', label='20.02.2023')
plt.plot(OTM_10Y1Y_2024, ten_y_1_y_2024.values, color='green',  label='20.02.2024')
plt.title('Swaption EUR 10Y1Y Skew', fontsize=18)
plt.xlabel('Strike', fontsize=14)
plt.ylabel('Implied Volatility', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', color='gray', linestyle='--', linewidth=0.5)
plt.legend(fontsize=12)
plt.savefig("/Users/nannaingemannohrt/Desktop/master_thesis/main/plots/10Y1Y.png")  # Save as PNG

# Display the plot
plt.show()
plt.close()
```
