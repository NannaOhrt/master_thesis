```python
import pandas as pd
from io import StringIO
import matplotlib.pyplot as plt
import yfinance as yf

#10Y10Y swaption
df = pd.read_excel('/Users/nannaingemannohrt/Desktop/master_thesis/main/data/ATM_10Y_10Y_from_2004_2024.xlsx',
                    sheet_name='Ark1', engine='openpyxl')

plt.figure(figsize=(10, 6))
plt.plot(df["Dato"],df["ATM_10Y_10Y"], color='grey', label= 'Swaption EUR 10Y10Y')
plt.xlabel('Year', fontsize=14)
plt.ylabel('Implied Volatility', fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid(axis='y', color='gray', linestyle='--', linewidth=0.5)
plt.legend(fontsize=14)
plt.savefig("/Users/nannaingemannohrt/Desktop/master_thesis/main/plots/10Y10Yswaption.png")
plt.show()
plt.close()

# returns 10Y10Y swaption
returns= df["ATM_10Y_10Y"].pct_change()
plt.figure(figsize=(10, 6))
plt.plot(df["Dato"], returns, label='Daily Returns Swaption EUR 10Y10Y', color='tab:blue')

plt.xlabel('Year', fontsize=14)
plt.ylabel('Returns', fontsize=14)
plt.legend(fontsize=14)
plt.grid(axis='y', color='gray', linestyle='--', linewidth=0.5)
plt.savefig("/Users/nannaingemannohrt/Desktop/master_thesis/main/plots/10Y10Yswaptionreturn.png")
plt.show()
plt.close()

#sp500
sp500 = yf.download('^GSPC', start='2004-01-01', end='2024-01-01')
sp500['Returns'] = sp500['Adj Close'].pct_change()
plt.figure(figsize=(10, 6))
plt.plot(sp500.index,sp500['Adj Close'], color='grey', label= 'S&P 500 (close values)')
plt.xlabel('Year', fontsize=14)
plt.ylabel('Close Values', fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid(axis='y', color='gray', linestyle='--', linewidth=0.5)
plt.legend(fontsize=14)
plt.savefig("/Users/nannaingemannohrt/Desktop/master_thesis/main/plots/sp500.png")
plt.show()
plt.close()

#sp500 returns
plt.figure(figsize=(10, 6))
plt.plot(sp500.index, sp500['Returns'], label='Daily Returns', color='tab:blue')
plt.xlabel('Date', fontsize=14)
plt.ylabel('Returns' ,fontsize=14)
plt.legend(fontsize=14)
plt.grid(axis='y', color='gray', linestyle='--', linewidth=0.5)
plt.savefig("/Users/nannaingemannohrt/Desktop/master_thesis/main/plots/sp500return.png")
plt.show()
plt.close()
```
