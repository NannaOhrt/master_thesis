```python
import yfinance as yf
import matplotlib.pyplot as plt

# Define the ticker symbol
tickerSymbol = 'AAPL'  # Example: Apple Inc.

# Get data on this ticker
tickerData = yf.Ticker(tickerSymbol)

# Get historical prices for this ticker
tickerDf = tickerData.history(period='1d', start='2010-1-1', end='2020-12-31')

# Plotting the closing prices
plt.figure(figsize=(10, 6))
plt.plot(tickerDf.index, tickerDf['Close'], label='Close Price')
plt.title(f'Closing Price of {tickerSymbol} (2010-2020)')
plt.xlabel('Date')
plt.ylabel('Close Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd

# Example ticker symbol
tickerSymbol = 'ACWI'

# Get data on this ticker
tickerData = yf.Ticker(tickerSymbol)

# Get historical prices for this ticker
tickerDf = tickerData.history(period='1d', start='2010-1-1', end='2020-12-31')

# Calculate the running maximum of the closing price to identify peaks
tickerDf['Running Max'] = tickerDf['Close'].cummax()

# Calculate drawdowns as the difference between the current price and the running maximum
tickerDf['Drawdown'] = tickerDf['Close'] - tickerDf['Running Max']

# Plotting the closing prices and drawdown periods
plt.figure(figsize=(14, 8))
plt.plot(tickerDf.index, tickerDf['Close'], label='Close Price', color='blue')
# Highlight drawdown periods
plt.fill_between(tickerDf.index, tickerDf['Close'], tickerDf['Running Max'], where=(tickerDf['Drawdown'] < 0), color='red', alpha=0.3, label='Drawdown Periods')
plt.title(f'Closing Price and Drawdown Periods of {tickerSymbol} (2010-2020)')

plt.xlabel('Date')
plt.ylabel('Close Price (USD)')
plt.legend()
plt.grid(True)

plt.show()
```
